# Implementation of Multivariate LR Batch GD

#### Import numpy, matplotlib.pyplot and make it inline

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read Data
file = "MultipleLR.csv"
Data2 = np.loadtxt(file, delimiter=',')
print(Data2.shape)
```

```python
# split input columns from data
x0 = np.ones(len(Data2)) # create ones column 
x_data = Data2[:,:3] # take first 3 columns from data as input 
x_data = np.insert(x_data, 0, x0, axis=1)  # add first column be ones column
x_data
```

```python
# split target column from data 
y_data = Data2[:,-1].reshape(-1, 1)
y_data.shape
```

#### Intialize Parameters

```python
# Intialize theta (we have 4 thetas so we define it in shape(4, 1))
theta = np.array([0, 0, 0, 0]).reshape(4, 1)
theta.shape
```

#### Predict y values using the LR equation
##### h(x)= theta_0 + theta_1 * x1 + theta_2 * x2 + theta_3 * x3

```python
def hypothesis(x_data, theta):
    # h(x) = X . theta   
    predict_Mul = x_data @ theta # multplication of matrices (25,4)x(4,1) = (25, 1)
    return predict_Mul
```

```python
predict_Mul = hypothesis(x_data, theta)
predict_Mul.shape
```

#### Calculate Cost Function 

```python
def cost_fun_mul(predict_Mul, y_data):
    # initialize m
    m = Data2.shape[0]   #return number of rows at data
    
    # calculate error (difference between predict output and target output) 
    error = predict_Mul - y_data

    # Calc. cost function (Loss fun.)
    J = (1/(2*m)) * (np.dot(error.T, error))  #   (1, 25) x (25, 1)

    return error, J
```

```python
error_Mul, J_Mul = cost_fun_mul(predict_Mul, y_data)
print(error_Mul.shape) #(25, 1)
print(J_Mul.shape) #(1, 1) 
```

#### Calculate rate of change of the parameters

```python
def Gradient_Mul(error_Mul):
    # initialize m
    m = Data2.shape[0]
    
    # Calc. Gredient for the parameters
    # Dim. (error_Mul.T ----> (1,25) ), (X_data -----> (25, 4) )
    gred_theta_mul = (1/m) * ((error_Mul.T @ x_data).reshape(4,1)) # Matrix Mul will return dim. (1, 4) so will use reshape 

    return gred_theta_mul
```

```python
gred_theta_mul= Gradient_Mul(error_Mul)
gred_theta_mul
```

#### Update parameters

```python
def update_Mul(theta, gred_theta_mul,Learning_rate):  
    # Update Parameters
    theta = theta - (Learning_rate * gred_theta_mul)
    
    # return updated parameters
    return theta
```

```python
theta_Mul = update_Mul(theta, gred_theta_mul, 0.0001)
```

#### Graph Function

```python
# function for plotting
def graph(x,y):
    fig = plt.subplots(figsize=(7,6))
    plt.plot(x, y, marker='o', c='blue', mfc='red', mec='k')
    plt.ylabel("Cost")
    plt.title("Learning curve of parameter and cost")
```

### Batch Gradient Descent for Multivariate

```python
# Initialization

theta = np.array([0, 0, 0, 0]).reshape(4, 1)
Learning_rate = 0.0001
num_iterations = 1000
thresold = 0.01
THETA_MUL =[]
Cost_J_Mul = []

counter = 0

#for loop about iterations
for i in range(num_iterations):
    # Calc. Y-hat
    predict_Mul = hypothesis(x_data, theta)
    
    #Calc error and cost 
    error_Mul, J_Mul = cost_fun_mul(predict_Mul, y_data)
    Cost_J_Mul.append(J_Mul[0,0])
    
    #Calc theta Gredient
    gred_theta_mul= Gradient_Mul(error_Mul)
    
    # Update thetas
    theta = update_Mul(theta, gred_theta_mul, Learning_rate)
    THETA_MUL.append(theta)
    
    counter += 1
    # check for early stoping
    if abs(Cost_J_Mul[i-1] - Cost_J_Mul[i]) < thresold and i > 1: # start from i > 1 check because before this we don't have past value  
        break
    print("Cost :",J_Mul,"  \nTheta :\n",theta)
    print('--------------------')


print("\n***************************************************************************************\n")
print("Model Will take",counter,"iterations to Converge")
```

```python
Cost_J_Mul = np.array(Cost_J_Mul)
Cost_J_Mul=  Cost_J_Mul.reshape(-1, 1)
Cost_J_Mul.shape
```

## LR Using sklearn

```python
# import Sklearn Library
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# read Data
file = "MultipleLR.csv"
Data4 = np.loadtxt(file, delimiter=',')
print(Data4)
```

```python
# extract x , y from data
X_sk_Multi = Data4[:,:3] 
Y_sk_Multi = Data4[:,-1]
```

```python
# Build LR Model
reg_Mul = LinearRegression()
```

```python
# fit the model to training data
reg_Mul.fit(X_sk_Multi, Y_sk_Multi)
```

```python
# Calc. Score of model
reg_Mul.score(X_sk_Multi, Y_sk_Multi)
```

```python
# Predict Y
Y_Predict_Mult = reg_Mul.predict(X_sk_Multi)
#Y_Predict_Mult
```

```python
# Calc accuracy by using r2 score
# to Calc. acuuracy
from sklearn.metrics import r2_score
r2_score(Y_sk_Multi, Y_Predict_Mult)
```
