## In this Notebook, I will develop a Python program that is able to implement the gradient descent variants(Batch  & Mini-Batch & Stochastic) in order to achieve the linear regression of a set of datapoints.
# -----------------------------------------------------------------------------------------------------------

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

#### I will generate a pair of arrays <b>X</b> and <b>y</b> with the values:
• <b>X</b> equally distributed between <b>0</b> and <b>20</b> 
• <b>y</b> such that: 
<b>yi = a*xi + b (and a = -2, b = 1)</b>


```python
# Create Data 
x = np.linspace(0, 20)
a = -2
b = 1
y = (a * x) + b
```

```python
print("X =\n",x)
print("--------"*10)
print("Y =\n",y)
```

## 1. Batch GD Function

```python
def Batch_GD(x, y, Learning_rate, num_iterations):
    # intialize theta_0 , theta_1
    theta_0 = 0
    theta_1 = 0

    # Num of samples in data
    m = len(x)
 
    # check value for early stopping
    thresold = 0.001
    
    # To count number of iteration will model take 
    counter= 0 
    
    # Lists to save values in them
    THETA_0 =[]
    THETA_1 =[]
    Cost_J = []
    Total_predict = [] 
    
    # Loop for max number of iterations
    for i in range(num_iterations):
        # Calc. Y-hat 
        predict = theta_0 + theta_1 * x
        Total_predict.append(predict)
            
        # calculate error (difference between predict output and target output) 
        error = predict - y

        # Calc. cost function (Loss fun.)
        J = (1/(2*m)) * (np.dot(error,error))
        Cost_J.append(J)
        
        THETA_0.append(theta_0)
        THETA_1.append(theta_1)
        
        print("Iteration :",(i+1),", Cost :",J,"  Theta_0 :",theta_0,"  Theta_1 :",theta_1)      
        
        # Calc. Gredient for the parameters
        gred_theta0 = (1/m) * sum(error)
        gred_theta1 = (1/m) * np.dot(error,x)

        # Update Parameters simultaneous 
        theta_0 = theta_0 - (Learning_rate * gred_theta0)
        theta_1 = theta_1 - (Learning_rate * gred_theta1)
    
        ###########
        # to plot all predict lines of regression created  
        ###########
#         fig, ax = plt.subplots(figsize=(7,6))
#         plt.plot(x, predict, 'r')
#         ax.scatter(x,y)
        ###########
        # increament counter of iterations
        counter += 1
        
        # check for early stoping
        if abs(Cost_J[i-1] - Cost_J[i]) < thresold and i > 1:
            # if condition achieved meaning that cost change with very very small value or not change 
            break
    
    print("\n***************************************************************************************\n")
    print("Model Will take",counter,"iterations to Converge")
    
    return THETA_0, THETA_1 , predict, Cost_J , Total_predict
```

### Try function 

```python
# Function ---> Batch_GD(x, y, Learning_rate, num_iterations)
theta_0, theta_1, predict, Cost_J, Total_predict = Batch_GD(x, y, 0.01, 100)
```

## 2. Stochastic GD Function

```python
def Stochastic_GD(x, y, Learning_rate, num_epochs):
    # intialize theta_0 , theta_1
    theta_0 = 0
    theta_1 = 0
    
    # check value for early stopping
    thresold = 0.001
    
    # Lists to save values in them
    THETA_0 =[]
    THETA_1 =[]
    Cost_J = []
    Total_predict = [] 
    
    # Loop for max number of epochs  
    for i in range(num_epochs):
        for j in range(len(x)): # iterations
            # Calc. Y-hat
            predict = theta_0 + theta_1 * x[j]

            # calculate error (difference between predict output and target output) 
            error = predict - y[j]

            # Calc. cost function (Loss fun.)
            J = (1/2) * (error*error)
            Cost_J.append(J)

            THETA_0.append(theta_0)
            THETA_1.append(theta_1)

            print("Epoch :",(i+1),", Iteration :",(j+1),", Cost :",J,"  Theta_0 :",theta_0,"  Theta_1 :",theta_1)   
            
            # Calc. Gredient for the parameters
            gred_theta0 = error
            gred_theta1 = error * x[j]

            # Update Parameters
            theta_0 = theta_0 - (Learning_rate * gred_theta0)
            theta_1 = theta_1 - (Learning_rate * gred_theta1)



        # check for early stoping
        if abs(Cost_J[i-1] - Cost_J[i]) < thresold and i > 1:
            break

    
    return THETA_0, THETA_1 , predict, Cost_J 
```

## 3.Mini-Batch GD Function

```python
# function to split data to mini-batches
def split_Data_Batches(x, y, size_of_Mini_Batch):
    # Split x , y for Batches depend on required size of batch 
    x_chunck = [x[i:i + size_of_Mini_Batch] for i in range(0, len(x), size_of_Mini_Batch)]
    y_chunck = [y[i:i + size_of_Mini_Batch] for i in range(0, len(y), size_of_Mini_Batch)]
    #print(x_chunck)
    #print(y_chunck)
    return x_chunck, y_chunck
```

```python
# Function ---> split_Data_Batches(x, y, size_of_Mini_Batch)
x_chunck, y_chunck = split_Data_Batches(x, y, 5)
#x_chunck
```

```python
def Mini_Batch_GD(x_chuncks, y_chuncks, Learning_rate, num_epochs):
    # intialize theta_0 , theta_1
    theta_0 = 0
    theta_1 = 0

    # intialize length of each mini batch (number elements )
    m = len(x_chuncks[0]) 
    
    # check value for early stopping
    thresold = 0.001
    
    THETA_0_MiniBatch =[]
    THETA_1_MiniBatch =[]
    Cost_J_MiniBatch = []
    Predict_Total = []
    
    # Loop for number of epochs
    for i in range(num_epochs):
        # iterations loop for each mini-batch created
        for j in range(len(x_chuncks)): 
            #print("X:",x_chuncks[j])
            #print("Y:",y_chuncks[j])
            
            # Calc. Y-hat
            predict = theta_0 + theta_1 * x_chuncks[j]
            Predict_Total.append(predict)
            
            # calculate error (difference between predict output and target output) 
            error = predict - y_chuncks[j]

            # Calc. cost function (Loss fun.)
            J = (1/(2*m)) * (np.dot(error, error))
            Cost_J_MiniBatch.append(J)

            THETA_0_MiniBatch.append(theta_0)
            THETA_1_MiniBatch.append(theta_1)

            print("Epoch :",(i+1),", Iteration :",(j+1),", Cost :",J,"  Theta_0 :",theta_0,"  Theta_1 :",theta_1)   
            
            # Calc. Gredient for the parameters
            gred_theta0 = (1/m) * (np.sum(error))
            gred_theta1 = (1/m) * (np.dot(error, x_chuncks[j]))

            # Update Parameters
            theta_0 = theta_0 - (Learning_rate * gred_theta0)
            theta_1 = theta_1 - (Learning_rate * gred_theta1)



        # check for early stoping
        if abs(Cost_J_MiniBatch[i-1] - Cost_J_MiniBatch[i]) < thresold and i > 1:
            # if condition achieved meaning that cost change with very very small value or not change 
            break

    
    return THETA_0_MiniBatch, THETA_1_MiniBatch, Predict_Total, Cost_J_MiniBatch 
```
