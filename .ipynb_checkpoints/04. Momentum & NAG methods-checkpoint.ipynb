## In this notebook, I will develop a Python program that is able to implement the accelerated gradient descent methods <b>(Momentum and NAG)</b> in order to achieve the linear regression of a set of datapoints.
# -----------------------------------------------------------------------------------------------------------

```python
import numpy as np
import matplotlib.pyplot as plt
```

I wil generate a pair of arrays <b>X</b> and <b>y</b> with the values in 
<b>X</b> equally distributed between <b>0</b> and 
<b>20</b> and the values in <b>y</b> such that: 
<b>yi = a*xi + b (and a = -1, b = 2)</b>


```python
# Create Data 
a = -1
b = 2
x = np.linspace(0, 20)
y = (a * x) + b
```

```python
print(x,"\n\n",y)
```

#### Plot your data points. 

```python
plt.scatter(x, y)
```

## â€¢ Batch GD

```python
def Batch_GD(x, y, Learning_rate, num_iterations):
    # intialize theta_0 , theta_1
    theta_0 = 0
    theta_1 = 0

    # intialize Learning rate
    #Learning_rate = 0.001
     
    m = len(x)
    #num_iterations = 100
    thresold = 0.001
    
    counter= 0 # to count number of iteration will model take 
    
    THETA_0 =[]
    THETA_1 =[]
    Cost_J = []
    
    Total_predict = []
    
    for i in range(num_iterations):
        # Calc. Y-hat
        predict = theta_0 + theta_1 * x
        Total_predict.append(predict)
            
        # calculate error (difference between predict output and target output) 
        error = predict - y

        # Calc. cost function (Loss fun.)
        J = (1/(2*m)) * (np.dot(error,error))
        Cost_J.append(J)
        
        THETA_0.append(theta_0)
        THETA_1.append(theta_1)
        
        print("Iteration :",(i+1),", Cost :",J,"  Theta_0 :",theta_0,"  Theta_1 :",theta_1)        
        # Calc. Gredient for the parameters
        gred_theta0 = (1/m) * sum(error)
        gred_theta1 = (1/m) * np.dot(error,x)

        # Update Parameters
        theta_0 = theta_0 - (Learning_rate * gred_theta0)
        theta_1 = theta_1 - (Learning_rate * gred_theta1)

        counter += 1
        # check for early stoping
        if abs(Cost_J[i-1] - Cost_J[i]) < thresold and i > 1:
            break
    
    print("\n***************************************************************************************\n")
    print("Model Will take",counter,"iterations to Converge")
    
    return THETA_0, THETA_1 , predict, Cost_J , Total_predict
```

```python
THETA_0, THETA_1 , predict, Cost_J , Total_predict = Batch_GD(x, y, 0.001, 100)
```

```python
plt.plot(x, predict, 'orange')
plt.scatter(x,y)
plt.title("Best Curve fit ")
```

```python
# Create Plotting Function
def graph(x,y):
    fig = plt.subplots(figsize=(7,6))
    plt.plot(x, y, marker='o')
    plt.ylabel("Cost")
```

```python
graph(THETA_0, Cost_J)
plt.xlabel("Theta (0)")
plt.title("Loss-Theta0")
```

```python
graph(THETA_1, Cost_J)
plt.xlabel("Theta (1)")
plt.title("Loss-Theta1")
```

```python
plt.plot(Cost_J)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Loss-Iterations")
```

```python
# to Calc. acuuracy
from sklearn.metrics import r2_score
r2_score(y, predict)
```

# 1. Momentum

### Update Batch GD for one variable implementation to be Momentum-Based GD and check results

```python
def Momentum_GD(x, y, Gamma, Learning_rate, num_iterations):
    # intialize theta_0 , theta_1
    theta_0 = 0
    theta_1 = 0
    
    # initalize exponentially weighted sum
    V_0 = 0
    V_1 = 0
    
    # initalize Momentum parameter 
    #Gamma = 0.1

    # intialize Learning rate
    #Alpha = 0.001
     
    m = len(x)

    threshold = 0.001
    
    counter= 0 # to count number of iteration will model take 
    
    THETA_0 =[]
    THETA_1 =[]
    Cost_J = []
    
    Total_predict = []
    
    for i in range(num_iterations):
        # Calc. Y-hat
        predict = theta_0 + (theta_1 * x)
        Total_predict.append(predict)
            
        # calculate error (difference between predict output and target output) 
        error = predict - y

        # Calc. cost function (Loss fun.)
        J = (1/(2*m)) * (np.dot(error,error))
        Cost_J.append(J)
        
        THETA_0.append(theta_0)
        THETA_1.append(theta_1)
        
        print("Iteration :",(i+1),", Cost :",J,"  Theta_0 :",theta_0,"  Theta_1 :",theta_1)        
        # Calc. Gredient for the parameters
        gred_theta0 = (1/m) * sum(error)
        gred_theta1 = (1/m) * np.dot(error,x)
        
        # Calc. the exponentially weighted sum 
        V_0 = (Gamma * V_0) + (Learning_rate * gred_theta0) 
        V_1 = (Gamma * V_1) + (Learning_rate * gred_theta1) 

        # Update Parameters
        theta_0 = theta_0 - V_0
        theta_1 = theta_1 - V_1

        counter += 1
        # check for early stoping
        if abs(Cost_J[i-1] - Cost_J[i]) < threshold and i > 1:
            break
    
    print("\n***************************************************************************************\n")
    print("Model Will take",counter,"iterations to Converge")
    
    return THETA_0, THETA_1 , predict, Cost_J , Total_predict
```

```python
# Function ---> Momentum_GD(x, y, Gamma, Learning_rate, num_iterations)
THETA_0_Momentum, THETA_1_Momentum , predict_Momentum, Cost_J_Momentum , Total_predict_Momentum = Momentum_GD(x, y, 0.5, 0.001, 100)
```

```python
plt.plot(x, predict_Momentum, 'orange')
plt.scatter(x,y)
plt.title("Best Curve fit ")
```

```python
graph(THETA_0_Momentum, Cost_J_Momentum)
plt.xlabel("Theta (0)")
plt.title("Loss-Theta0")
```

```python
graph(THETA_1_Momentum, Cost_J_Momentum)
plt.xlabel("Theta (1)")
plt.title("Loss-Theta1")
```

```python
plt.plot(Cost_J_Momentum)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Loss-Iterations")
```

```python
# to Calc. acuuracy
from sklearn.metrics import r2_score
r2_score(y, predict_Momentum)
```

# 2. NAG

### Update your Batch GD for one variable  implementation to be NAG and check your results

```python
def NAG(x, y, Gamma, Learning_rate, num_iterations):
    # intialize theta_0 , theta_1
    theta_0 = 0
    theta_1 = 0
    
    # initalize exponentially weighted sum
    V_0 = 0
    V_1 = 0
    
    # initalize Momentum parameter 
    #Gamma = 0.1

    # intialize Learning rate
    #Alpha = 0.001
     
    m = len(x)

    threshold = 0.001
    
    counter= 0 # to count number of iteration will model take 
    
    THETA_0 =[]
    THETA_1 =[]
    Cost_J = []
    
    Total_predict = []
    
    for i in range(num_iterations):
        # Calc. Y-hat
        predict = theta_0 + (theta_1 * x)
        Total_predict.append(predict)
            
        # calculate error (difference between predict output and target output) 
        error = predict - y

        # Calc. cost function (Loss fun.)
        J = (1/(2*m)) * (np.dot(error,error))
        Cost_J.append(J)
        
        THETA_0.append(theta_0)
        THETA_1.append(theta_1)
        
        print("Iteration :",(i+1),", Cost :",J,"  Theta_0 :",theta_0,"  Theta_1 :",theta_1)        
        
        # Calc. Temp. thetas 
        theta_temp_0 = theta_0 - (Gamma * V_0)
        theta_temp_1 = theta_1 - (Gamma * V_1)
        
        # Calc. temp. Hypothesis 
        H_temp = theta_temp_0 + (theta_temp_1 * x)
        
        # Calc. difference between H_temp , real out y
        error_temp = H_temp - y
        
        # Calc. Gredient for temp. parameters
        gred_theta_temp_0 = (1/m) * sum(error_temp)
        gred_theta_temp_1 = (1/m) * (np.dot(error_temp, x))
        
        # Update Parameters
        theta_0 = theta_temp_0 - (Learning_rate * gred_theta_temp_0)
        theta_1 = theta_temp_1 - (Learning_rate * gred_theta_temp_1)
        
        # Calc. the exponentially weighted sum 
        V_0 = (Gamma * V_0) + (Learning_rate * gred_theta_temp_0) 
        V_1 = (Gamma * V_1) + (Learning_rate * gred_theta_temp_1) 


        counter += 1
        # check for early stoping
        if abs(Cost_J[i-1] - Cost_J[i]) < threshold and i > 1:
            break
    
    print("\n***************************************************************************************\n")
    print("Model Will take",counter,"iterations to Converge")
    
    return THETA_0, THETA_1 , predict, Cost_J , Total_predict
```

```python
THETA_0_NAG, THETA_1_NAG , predict_NAG, Cost_J_NAG , Total_predict_NAG = NAG(x, y, 0.5, 0.001, 100)
```

```python
plt.plot(x, predict_NAG, 'orange')
plt.scatter(x,y)
plt.title("Best Curve fit ")
```

```python
graph(THETA_0_NAG, Cost_J_NAG)
plt.xlabel("Theta (0)")
plt.title("Loss-Theta0")
```
